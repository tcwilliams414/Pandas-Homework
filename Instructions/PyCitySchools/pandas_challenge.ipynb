```python
#Scores by School Spending avg math, avg reading, % passing math, % passing reading, overall passing rate
labels= ["<$585","$585-630","$630-645","$645-680"]
bins= [0,584.999,630.999,645.99,680.99]
merged_df["spending_bins"]=pd.cut(merged_df["budget"]/merged_df["size"], bins, labels=labels)
```

```python
spending_df= merged_df.groupby("spending_bins")

math_spending= spending_df["math_score"].mean()
reading_spending= spending_df["reading_score"].mean()
pass_math_spending= merged_df[merged_df["math_score"] >=70].groupby("spending_bins")["Student ID"].count()/spending_df["Student ID"].count()*100
pass_reading_spending= merged_df[merged_df["reading_score"] >=70].groupby("spending_bins")["Student ID"].count()/spending_df["Student ID"].count()*100
overall_spending= merged_df[(merged_df["reading_score"] >=70) & (merged_df["math_score"] >=70)].groupby("spending_bins")["Student ID"].count()/spending_df["Student ID"].count()*100
```

```python
spending_by_school_df= pd.DataFrame({
    "Average Math Score": math_spending,
    "Average Reading Score": reading_spending,
    "% Passing Math": pass_math_spending,
    "% Passing Reading": pass_reading_spending, 
    "Overall Passing ": overall_spending 
})
spending_by_school_df
```

```python
#Scores by School Size
school_labels= ["Small(<1000)", "Medium (1000-2000)", "Large(2000-5000)"]
school_bins= [0,1000,2000,5000]
merged_df["size_bins"]=pd.cut(merged_df["size"],school_bins, labels=school_labels)
```

```python
size_df= merged_df.groupby("size_bins")
```

```python
math_size= size_df["math_score"].mean()
reading_size= size_df["reading_score"].mean()
pass_math_size= merged_df[merged_df["math_score"] >=70].groupby("size_bins")["Student ID"].count()/size_df["Student ID"].count()*100
pass_reading_size= merged_df[merged_df["reading_score"] >=70].groupby("size_bins")["Student ID"].count()/size_df["Student ID"].count()*100
overall_size= merged_df[(merged_df["reading_score"] >=70) & (merged_df["math_score"] >=70)].groupby("size_bins")["Student ID"].count()/size_df["Student ID"].count()*100
```

```python
#Data Frame
scores_by_school_size= pd.DataFrame({
    "Average Math Score": math_size,
    "Average Reading Score":reading_size,
    "% Passing Math":pass_math_size,
    "% Passing Reading":pass_reading_size,
    "Overall Passing":overall_size   
})

scores_by_school_size
```

```python
# Scores by School Type
scores_by_school_type_df= merged_df.groupby("type")

school_math=scores_by_school_type_df["math_score"].mean()
school_reading= scores_by_school_type_df["reading_score"].mean()
school_pass_math= merged_df[merged_df["math_score"]>=70].groupby("type")["Student ID"].count()/scores_by_school_type_df["Student ID"].count()*100
school_pass_reading= merged_df[merged_df["reading_score"]>=70].groupby("type")["Student ID"].count()/scores_by_school_type_df["Student ID"].count()*100
school_overall= merged_df[(merged_df["math_score"]>=70) &(merged_df["reading_score"] >=70)].groupby("type")["Student ID"].count()/scores_by_school_type_df["Student ID"].count()*100                               
```

```python
scores_by_school_type= pd.DataFrame({
    "Average Math Score": school_math,
    "Average Reading Score": school_reading,
    "% Passing Math":school_pass_math,
    "% Passing Reading":school_pass_reading,
    "Overall Passing":school_overall   
})

scores_by_school_type
```
